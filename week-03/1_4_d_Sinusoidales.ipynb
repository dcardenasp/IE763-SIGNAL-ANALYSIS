
# Características de las señales sinusoidales análogas y discretas en cuanto a la frecuencia.

Ya que el concepto de frecuencia es conocido y sabemos que su dimensión es inversa a la del tiempo, modificaciones al argumento tiempo de una función (continuo o discreto) afectará la naturaleza de la frecuencia.

# Señales sinusoidales continuas en el tiempo

Una oscilación armónica simple se describe matemáticamente mediante la señal sinusoidal
 
$$
x(t) = Acos(\Omega t + \theta)
$$
 
La señal sinusoidal se caracteriza por tres parámetros:
 
* $A$: Amplitud
* $\Omega$: Frecuecnia en $rad/seg$
* $\theta$: Fase en $rad$

Donde $\Omega = 2\pi F$, con $F$ dado en $ciclos/seg$ ó $Hz$.
 

En la siguiente figura se muestra la gráfica de la señal $x_a(t) = 5\cdot Cos\left(0.5 \pi t + \frac{\pi}{3}\right)$


```
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
```

```
t = np.arange(-2, 12, .05)
x = 5*np.cos(0.5*np.pi*t + np.pi/3)
plt.figure(figsize=(25, 10))
plt.plot(t, x)
plt.grid()
plt.margins(x=0)
plt.title('Gráfica señal $x_a(t) = 5\cdot Cos(0.5 \pi t + \pi/3)$', fontsize=15)
plt.xlabel('Tiempo')
an1 = plt.annotate("",xy=(10/3, 5), xycoords='data', xytext=(5, 0.8), textcoords='data', size=20, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"), arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=0.2",
                                  relpos=(0., 0.), fc="w"))
an1 = plt.annotate("Amplitud", xy=(10/3, 0), xycoords='data', xytext=(5, 0.8), textcoords='data', size=20, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"), arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=-0.2",
                                  relpos=(1., 0.), fc="w"))
an2 = plt.annotate("",xy=(22/3, 0), xycoords='data', xytext=(28/3, 0), textcoords='data', size=20, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"), arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=0.2",
                                  relpos=(0., 0.), fc="w"))
an2 = plt.annotate("Periodo\n Fundamental", xy=(34/3, 0), xycoords='data', xytext=(28/3, 0), textcoords='data', size=20, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"), arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=-0.2",
                                  relpos=(0., 0.), fc="w"))
an3 = plt.annotate("",xy=(0, 0), xycoords='data', xytext=(-np.pi/6, -2), textcoords='data', size=20, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"), arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=0.2",
                                  relpos=(0., 0.), fc="w"))
an3 = plt.annotate("Fase", xy=(-1/6*4, 0), xycoords='data', xytext=(-np.pi/6, -2), textcoords='data', size=20, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"), arrowprops=dict(arrowstyle="-|>", connectionstyle="arc3,rad=-0.2",
                                  relpos=(0., 0.), fc="w"))
plt.show()
```

La siguiente figura interactiva permite variar los parámetros de amplitud, frecuencia y fase de la señal $x(t) = ASen(2\pi ft + \theta)$.

```
import ipywidgets
def seno(A, f, theta):
  t = np.arange(0, 10, .001)
  x = A*np.sin(2*np.pi*f*t + theta)
  plt.figure(figsize=(25, 10))
  plt.plot(t, x)
  plt.grid()
  plt.margins(x=0)
  plt.title('Gráfica señal $x(t) = {}\cdot Sen(2\pi \cdot {}\cdot t + {})$'.format(A, f, theta), fontsize=15)
  plt.xlabel('Tiempo')
  plt.ylim(-10, 10)

display(ipywidgets.interactive(seno, A=ipywidgets.FloatSlider(value=1, min=0, max=20, step=.1, description='Amplitud: ', continuous_update=False),
                               f=ipywidgets.FloatLogSlider(value=1, min=-3, max=2, description='Frecuencia: ', continuous_update=False),
                               theta=ipywidgets.FloatSlider(value=0, min=0, max=2*np.pi, step=np.pi/10, description='Fase: ', continuous_update=False)))
```

# Propiedades de las señales sinusoidales con relación a la frecuencia

1. Para todo valor fijo de frecuencia $F$, $x(t)$ es siempre periodica.
 
$$
x(t) = x(t+T_0)
$$

>Donde $T_0$ es el periodo fundamental de la señal.

2. Las señales sinusoidales continuas en el tiempo de diferentes frecuencias son diferentes.

3. Un incremento en la frecuencia $F$, generará un incremento en las oscilaciones de la señal, es decir, en un mismo intervalo de tiempo se presentarán un mayor número de ciclos.

Dado lo anterior, estas características son también extensibles a las señales exponenciales complejas.

$$
x(t) = A e^{j~(\Omega t + \theta)}
$$

Donde

$$
e^{\pm j~\phi} = Cos(\phi) \pm j~Sen(\phi)
$$

Por lo tanto, la señal $x(t) =  Acos(\Omega t + \theta)$ puede definirse de la siguiente manera.

$$
x(t) =  Acos(\Omega t + \theta) = \frac{A}{2}e^{j~(\Omega t + \theta)} + \frac{A}{2}e^{-j~(\Omega t + \theta)}
$$

La siguiente figura interactiva permite visualizar la señal $x(t) = Cos(2\pi t)$ como suma de dos fasores en el plano complejo y como función del tiempo.

```
def animation(t0=0.1):
  x_1 = .5*np.exp(1j*2*np.pi*t0)
  x_2 = np.conj(x_1)
  x = x_1 + x_2
  fig = plt.figure(figsize=(25, 10))
  ax1 = fig.add_axes([0, 0, 0.5, 0.8], projection='polar')
  ax1.set_rmax(1)
  ax1.set_ylabel('Eje Real', labelpad=30)
  ax1.set_xlabel('Eje Imaginario')
  arr_x1 = plt.arrow(np.angle(x_1), 0, 0, abs(x_1), width = 0.015, length_includes_head = True,
                    edgecolor = 'blue', facecolor = 'blue', lw = 2)
  arr_x2 = plt.arrow(np.angle(x_2), 0, 0, abs(x_2), width = 0.015, length_includes_head = True,
                    edgecolor = 'red', facecolor = 'red', lw = 2)
  arr_x = plt.arrow(np.angle(x), 0, 0, abs(x), width = 0.015, edgecolor = 'green', length_includes_head = True, facecolor = 'green')
  ax2 = fig.add_axes([0.5, 0, 0.5, 0.8])
  ax2.grid()
  t = np.arange(0, t0+.01, .01)
  x = .5*np.exp(1j*2*np.pi*t)
  ax2.plot(t, np.real(x + np.conj(x)))
  ax2.set_xlim(0, 1)
  ax2.set_ylim(-1, 1)
  ax2.set_ylabel('Eje Real')
  ax2.set_xlabel('Tiempo')
  fig.show()
  fig.suptitle('Gráfica señal $x(t) = Cos(2\pi t)$', fontsize=20)
display(ipywidgets.interactive(animation, t0=ipywidgets.FloatSlider(value=0.1, step=.01, max = 1, min = 0, description='Valor de t:'
, continuous_update=False)))
```

## **Señales Sinusoidales Discretas En El Tiempo**

Una señal discreta se puede expresar de la siguiente forma;
 
Dada una señal análoga:
 
$$
x(t) = ACos(\Omega t + \theta)
$$
 
Donde el argumento t, es discretizado a partir de “$n\cdot T_s$”; donde $n$ representa la enésima muestra y $T_s$ el tiempo fijo de la toma entre muestras, es decir, el periodo de muestreo, se tiene:
 
 
$$
x(n\cdot T_S) = ACos(\Omega \cdot(n\cdot T_s) + \theta)\\
x(n\cdot T_S) = ACos(2\pi \cdot F \cdot T_s \cdot n + \theta)
$$
 
Donde $T_s = \frac{1}{F_s}$
 
$$
x(n\cdot T_S) = ACos\left(2\pi \cdot \frac{F}{F_s} \cdot n + \theta\right)
$$
 
Con $\frac{F}{F_s} = f$.
 
La relación entre la frecuencia de la señal $F$ y la frecuencia de muestreo $F_s$, se denomina frecuencia discreta o frecuencia normalizada. Por lo tanto la señal sinusoidal discreta se denota por:
 
$$
x(n\cdot T_S) = ACos(2\pi \cdot f \cdot n + \theta)
$$
 
El término $T_s$ en la notación de $x(nT_s)$, generalmente no se escribe, quedando solamente $x(n)$, por lo tanto, la expresión quedará

  
$$
x(n) = ACos(2\pi \cdot f \cdot n + \theta)
$$
 

En este punto se tienen tres tipos de frecuencias que deben ser tenidas en cuenta:

* $F$: frecuencia de la señal análoga, dada en $\frac{Ciclos}{Seg}$.
* $F_s$: frecuencia de muestreo, dada en  $\frac{Muestras}{Seg}$.
* $f$: frecuencia discreta, dada en las siguientes unidades;
$$
\frac{\frac{Ciclos}{Seg}}{\frac{Muestras}{Seg}} = \frac{Ciclos}{Muestras}
$$


### Ejemplo 1

La señal $Cos(\frac{\pi}{6}n+\frac{\pi}{3})$ es igual a $Cos(2 \pi \frac{1}{12}n+\frac{\pi}{3})$. La frecuencia es de $\frac{1}{12}$, es decir, cada ciclo contiene 12 muestras.

```
def animation(k):

  plt.figure(figsize=(25, 10))
  n = np.arange(0, 37)
  x = np.cos(2*np.pi*k/12 *n + np.pi/3)
  plt.stem(n, x, use_line_collection=True)
  plt.margins(x=0.01)
  plt.title('Gráfica de la señal $x[n] = Cos(2\pi k n/12 + \pi/3)$', fontsize=15)
  plt.xlabel('Muestras')
  plt.ylim([-1.1,1.1])
  plt.xticks(n)
  plt.grid()
  plt.show()

display(ipywidgets.interactive(animation, k=ipywidgets.FloatSlider(value=1, step=1, max = 20, min = 1, description='Valor de k:'
  , continuous_update=False)))
```

### Propiedades de las señales discretas en la frecuencia.

1. Una señal sinusoidal de tiempo discreto es periódica, sólo si su frecuencia discreta es un número racional, es decir el cociente de dos números enteros.
 
$$
x[n] = x(n+N)~~~ \forall ~n~y~  N \in \mathbb{Z}
$$
 
Donde $N$ es el mínimo valor entero para que la relación anterior se cumpla y toma el nombre de periodo fundamental.
 
### Ejemplo 2
 
Dada la señal $x[n] = Cos(2\pi \cdot f \cdot n)$, demostrar que es una señal periódica.
 
$$
Cos(2\pi \cdot f \cdot n) = Cos(2\pi \cdot f \cdot (n+N))\\
Cos(2\pi \cdot f \cdot n) = Cos(2\pi \cdot f \cdot n)\cdot Cos(2\pi \cdot f \cdot N) - Sen(2\pi \cdot f \cdot n)\cdot Sen(2\pi \cdot f \cdot N)   
$$
 
Para que se cumpla la igualdad (condición de periodicidad), $Cos(2\pi \cdot f \cdot N) = 1$ y $Sen(2\pi \cdot f \cdot N) = 0$. Esto solo ocurre cuando
 
$$
2\pi \cdot f \cdot N = 2\pi \cdot k,~~~\forall~ k \in \mathbb{N}
$$
 
Por lo tanto, la frecuencia discreta es la siguiente:
 
$$
f = \frac{k}{N}
$$
 
Como la relación $k/N$ es de dos números enteros, $f$ deberá ser un número racional para que la señal $x[n]$ sea periódica. 
 
La relación $f$ deberá reducir hasta que los valores de $k$ y $N$ sean primos relativos.
 
Por ejemplo para una frecuencia discreta 2/12, se cuenta con un periodo fundamental de 6 muestras que conformarán un ciclo.
 
$$
f = \frac{2}{12} = \frac{1}{6}
$$
 

2. Señales sinusoidales discretas en el tiempo cuyas frecuencias están separadas múltiplos enteros de $2\pi$ son idénticas.
 
Si se toma por ejemplo la señal   $x(n)=Cos(w\cdot n)$, y modifico su frecuencia angular en $(w+2\pi\cdot L)]$, es decir en múltiplos enteros de $2\pi$L, donde $L$ es un entero, entonces x(n) seguirá siendo igual.

$$
Cos(w\cdot n) = Cos[(w+2\pi\cdot L)\cdot n]\\
Cos(w\cdot n) = Cos(w\cdot n) \cdot Cos(2\pi\cdot L\cdot n) - Sen(w\cdot n) \cdot Sen(2\pi\cdot L\cdot n)
$$


 
Con $L$ y $n$  números enteros, su producto, suma o diferencia será otro número entero.
 
En consecuencia, todas las secuencias
 
$$
x_L(n) = ACos(w_L n)\\
w_L = w_0 + 2\pi L, ~~~ -\pi \leq w_0 < \pi
$$
 
son identicas. Los valores de frecuencia tales que $|w|>\pi$ se llaman *aliasing*.



```
import matplotlib as mpl
fig, ax = plt.subplots(figsize=(25, 5))
fig.subplots_adjust(bottom=0.5)
cmap = mpl.cm.cool
bounds = [-2*np.pi, -np.pi, np.pi, 2*np.pi]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
cb = mpl.colorbar.ColorbarBase(ax, cmap=cmap, norm=norm, orientation='horizontal')
cb.ax.set_xticklabels(['$-2 \pi$', '$-\pi$', '$\pi$', '$2 \pi$'], fontsize=30)
cb.set_label('$w_0$', fontsize=20)
plt.text(0, 0.5, 'Frecuencias Originales \n $-\pi \leq w_0 < \pi$', size=35, ha="center", va="center")
plt.text(-1.35*np.pi, 0.5, 'Frecuencias Aliasing \n $ w_0 < -\pi$', size=35, ha="center", va="center")
plt.text(1.35*np.pi, 0.5, 'Frecuencias Aliasing  \n $\pi \leq w_0 $', size=35, ha="center", va="center")
fig.show()
```

### Ejemplo 3

Sea la señal $x(n) = Cos(\frac{\pi}{6}n)$. Determine si es una señal original o es un aliasing.

$$
w_0 = \frac{\pi}{6} < \pi 
$$

Por lo tanto, la señal $x(n)$ es una señal original.


```
plt.figure(figsize=(25, 10))
n = np.arange(0, 37)
x = np.cos(np.pi/6 *n)
plt.stem(n, x, use_line_collection=True)
plt.margins(x=0.01)
plt.title('Gráfica de la señal $x[n] = Cos(n \cdot \pi/6)$', fontsize=15)
plt.xlabel('Muestras')
plt.xticks(n)
plt.grid()
plt.show()
```

### Ejemplo 4
Determine si la señal $x[n] = Cos(\frac{4 \pi}{3} n)$ es una señal original o un aliasing.

$$
w = \frac{4 \pi}{3} > \pi
$$
Por lo tanto, la señal $x[n]$ es un aliasing de otra señal de menor frecuencia. Como esto ocurre para señales separadas múltiplos de $2 \pi$, bastará con determinar cual es el grado de separación en múltiplos de $2 \pi L$ sobre la señal con aliasing para hallar la original.

$$
w_0 = \frac{4 \pi}{3} - 2 \pi   = -\frac{2 \pi}{3}
$$
La señal discreta original es $ Cos(-\frac{2 \pi}{3} n) =  Cos(\frac{2 \pi}{3} n)$. La señal discreta original es $ Cos(-\frac{2 \pi}{3} n) =  Cos(\frac{2 \pi}{3} n)$. 


```
n = np.arange(0, 16)
t = np.linspace(0, 15, 1000)
x_1 = np.cos(4*np.pi/3 *n)
x1 = np.cos(4*np.pi/3 *t)
x_2 = np.cos(-2*np.pi/3 *n)
x2 = np.cos(-2*np.pi/3 *t)
fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(25, 10))
ax1.stem(n, x_1, use_line_collection=True, linefmt='grey', markerfmt='D')
ax1.plot(t, x1, ls=":")
ax2.stem(n, x_2, use_line_collection=True, linefmt='grey', markerfmt='D')
ax2.plot(t, x2, ls=":")
ax1.set_title(r'Gráfica de la señal $x[n] = Cos( \frac{4 \pi}{3} n)$', fontsize=15)
ax2.set_title(r'Gráfica de la señal $x[n] = Cos( \frac{2 \pi}{3} n)$', fontsize=15)
ax2.set_xlabel('Muestras')
ax1.grid()
ax2.grid()
ax1.margins(x=0.01)
ax2.margins(x=0.01)
fig.show()
```

De la propiedad anterior se deriva que la frecuencia de muestreo debe ser mayor que el doble de la frecuencia de la señal para que no exista aliasing o copias. Esto se conoce como teorema de muestreo o teorema de Niquiz.

$$
|w| < \pi \\
|2 \pi f| < \pi \\
|f| < \frac{1}{2}\\
|\frac{F}{F_s}| < \frac{1}{2}\\
2|F| < |F_s|
$$



3. La tasa de oscilación más alta de una señal sinusoidal discreta en el tiempo, se alcanza cuando $|w| = \pi$, o lo que es lo mismo $|f| = \frac{1}{2}$.

La siguiente figura interactiva permite evaluar la función $x[n] = Cos(2 \pi f n)$ en una secuencia de frecuencias dadas.

```
def senial(f):
  n = np.arange(0, 50)
  plt.figure(1, (25, 10))
  x = np.cos(2*np.pi*f*n)
  plt.stem(n, x, use_line_collection=True)
  plt.grid()
  plt.margins(x=.01)
  plt.title('Gráfica de la señal $x[n] = cos(2\pi \cdot {} \cdot n)$'.format(f), fontsize=15)
  plt.xlabel('Muestras')
  plt.show()
ipywidgets.interact(senial, f=[('0', 0), ('1/16', 1/16), ('1/8', 1/8), ('1/4', 1/4), ('1/2', 1/2)])
```

Es interesante también saber, aunque ya se sepa, que ocurre entre $\pi < w_0 < 2 \pi$. Para ello consideremos sinusoides de frecuencia $w_1 = w_0$ y $w_2 = 2\pi – w_0$. Note que $w_1$ varía entre $\pi$ y $2 \pi$ y $w_2$ varía entre $\pi$ y $0$.

$$
x_1(n) = ACos(w_1n) = ACos(w_0n)\\
x_2(n) = ACos(w_2n) = ACos((2 \pi -w_0)n) = ACos(-w_0n)\\
x_2(n) = x_1(n)
$$

Por lo tanto, $x_1(n)$ es un aliasing de $x_2(n)$.




```
fig = plt.figure(figsize=(25, 5))
ax = fig.add_subplot(111)
ax.set_xlim(-7,7)
ax.set_ylim(3,7)
plt.hlines(5, -7, 7)
plt.hlines(5, 0, np.pi, colors='yellow').set_linewidth(5)
plt.hlines(5, np.pi, 2*np.pi, colors='green').set_linewidth(5)
for i, j in zip(np.arange(-2*np.pi, 3*np.pi, np.pi), ['$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$']):
  plt.vlines(i, 4.5, 5.5)
  plt.text(i, 4, j, fontsize = 20, horizontalalignment='center')
plt.annotate('$\omega_2$', (np.pi/2,5), xytext = (np.pi/2 - 1, 6), fontsize = 30,  
                 arrowprops=dict(facecolor='black', shrink=0.1), horizontalalignment='right')
plt.annotate('$\omega_1$', (np.pi/2+np.pi,5), xytext = (np.pi/2+np.pi - 1, 6), fontsize = 30,  
                 arrowprops=dict(facecolor='black', shrink=0.1), horizontalalignment='right')
plt.axis('off')
plt.show()
```

## Resumen
Una señal análoga continua en el tiempo está dada por:

$$
x(t) = ACos(\Omega t + \theta)
$$
Y una señal discreta, con $F_s = 1/T_S$ está dada por:
$$
x[n] = ACos(2 \pi f n + \theta)
$$
Donde $f = \frac{F}{F_s}$ y $\omega = \Omega \cdot T_s$.

Entonces, los rangos de frecuencia continua son:

$$
-\infty < F < \infty\\
-\infty < \Omega < \infty
$$

Y para frecuencia discreta:

$$
-\frac{1}{2} \leq f < \frac{1}{2}\\
-\pi \leq \omega < \pi
$$


```
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.text(0.5, 0.92, 'Tabla resumen', fontsize = 25, horizontalalignment='center')
ax.text(0.25, 0.75, 'Señales continuas\n en el tiempo', fontsize = 20, horizontalalignment='center')
ax.text(0.75, 0.75, 'Señales discretas\n en el tiempo', fontsize = 20, horizontalalignment='center')
ax.text(0.25, 0.6, '$\Omega = 2\pi \cdot F$'+'\n'+r'$\frac{Rad}{Seg} ~~~~~ \frac{Ciclos}{Seg}$', fontsize = 20, horizontalalignment='center')
ax.text(0.75, 0.6, '$\omega = 2\pi \cdot f$'+'\n'+ r'$\frac{Rad}{Muestra} ~~~~~ \frac{Ciclos}{Muestra}$', fontsize = 20, horizontalalignment='center')
bbox_props_1 = dict(boxstyle="rarrow,pad=0.3", fc="cyan", ec="b", lw=2)
ax.text(0.5, 0.4, '$\omega = \Omega \cdot T_s$'+'\n'+'$f = F/F_s$', fontsize = 20, horizontalalignment='center', bbox=bbox_props_1)
ax.text(0.25, 0.4, '$-\infty < \Omega < \infty$'+'\n'+'$-\infty < F < \infty$', fontsize = 20, horizontalalignment='center')
ax.text(0.75, 0.4, '$-\pi \leq \omega < \pi$'+'\n'+r'$-\frac{1}{2} \leq f < \frac{1}{2}$', fontsize = 20, horizontalalignment='center')
bbox_props_2 = dict(boxstyle="larrow,pad=0.3", fc="cyan", ec="b", lw=2)
ax.text(0.5, 0.1, '$\Omega = \omega / T_s$'+'\n'+'$F = f \cdot F_s$', fontsize = 20, horizontalalignment='center', bbox=bbox_props_2)
ax.text(0.75, 0.1, r'$-\frac{\pi}{T_s} \leq \Omega < \frac{\pi}{T_s}$'+'\n'+r'$-\frac{F_s}{2} \leq F < \frac{F_s}{2}$', fontsize = 20, horizontalalignment='center')
ax.set_xticks([])
ax.set_yticks([])
ax.set_facecolor('#eafff5')
fig.show()
```
