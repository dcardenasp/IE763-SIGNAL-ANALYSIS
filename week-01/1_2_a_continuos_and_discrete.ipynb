```
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

#	A. Señales Continuas y discretas



*   Una señal $x(t)$ es continua si su argumento es una variable continua, $t\in\mathbb{R}$. 
*   Una señal $x[n]$ es discreta si su argumento es una variable discreta, $n\in\mathbb{N}$.


1.   A continuación se simula una señal continua a través de sympy 


```
t = sym.Symbol('t')
x = 10*sym.sin(t)+4*sym.sin(2*t)-6*sym.sin(3*t)
p1=sym.plotting.plot(x,(t, 0, 6),xlim=[0,6],ylim=[-20,20],xlabel='t',ylabel='x(t)')
```


2.   A continuación se crea una función de la cual se toman muestras en tiempo para construir una señal discreta

```
#Definición de la función:
f = lambda a : 10* np.sin(a)+4*np.sin(2*a)-6*np.sin(3*a)

#Definición de los instantes de muestreo:
t_n = np.array([0,0.1,0.5,1.0,1.1,1.2,1.3,1.4,1.8,2.0,2.8,3.5,4,4.6,5])

#Muestreo de la función:
x = f(t_n)
print('Tiempos',t_n.shape,'Muestras',x.shape)
#Gráfica
plt.stem(t_n,x)
plt.ylim([-20,20])
plt.xlim([0,6])
plt.show()
```

```
#Periodo de muestreo
Ts = 0.1 #segundos
#Variable discreta independiente
n = np.arange(1000) #adimensional
#Tiempos de muestreo
t_n = n*Ts #segundos

#for i,j in zip(n,t_n):
#  print(i,j)
  
#Muestreo
x = f(t_n)
#Gráfica
plt.stem(t_n,x)
plt.ylim([-20,20])
plt.xlim([0,6])
plt.show()
```
