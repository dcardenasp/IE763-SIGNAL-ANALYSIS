# **2. Señales De Energía Y Señales De Potencia**

<p align = "justify"><b>Alcance: </b>Se definen las señales de potencia y energia y sus principales características.

<p align = "justify">Las señales pueden clasificarse como de energía o de potencia. Los conceptos de energía y potencia son empleados para la generalización de métricas que permitirán tomar decisiones de convergencia en distintos tipos de análisis en espacios de representación.

## **2.1. Señales de energía**

<p align = "justify">Se dice que una señal es de energía, si su energía total $E_x$ converge, es decir:

\begin{equation}
0< E_x < \infty \tag{2.1}
\end{equation}

<p align = "justify">Para el caso de una señal $x(t)$ contínua, su energía es calculada como:

\begin{equation}
E_x = \|x(t) \|^2 = \int_{-\infty}^{\infty} |x(t)|^{2}~dt \tag{2.2}
\end{equation}

<p align = "justify">donde la norma de la señal $x(t)$ se denota por $\|x(t) \|$. En el mismo sentido, la energía de la función discreta $x[n]$, es:

\begin{equation}
E_x = \|x[n] \|^2 = \sum_{n=-\infty}^{\infty} |x[n]|^{2} \tag{2.3}
\end{equation}

<p align = "justify">Si se define la energía de la señal en un intervalo de tiempo $T$ como:

\begin{equation}
E_T = \int_T |x(t)|^{2}~dt \tag{2.4}
\end{equation}

<p align = "justify">Su energía total puede expresarse en términos del siguiente límite:

\begin{equation}
E_x = \lim_{T \to \infty} E_T \tag{2.5}
\end{equation}

<p align = "justify">Para las señales discretas:

\begin{equation}
E_N = \sum_{n=-N}^{N} |x[n]|^{2} \tag{2.6}
\end{equation}

<p align = "justify">Y la energía total de $x[n]$ es:

\begin{equation}
E_x = \lim_{N \to \infty} E_N \tag{2.7}
\end{equation}


Código 2.1: Cálculo de la energía de una señal sinusoidal.

```
# Se llaman las librerias a usar
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as ipy


# Se crea una función para crear gráficos personalizados
def Plot(*arg, label=''):

    fig, ax= plt.subplots(1, figsize=(15, 8))
    ax.plot(*arg, lw=3)
    ax.spines['left'].set_position(('data',0))
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['top'].set_color('none')
    ax.set_axisbelow(False)

    for i in ax.get_xticklabels() + ax.get_yticklabels():
        i.set_fontsize(15)
        i.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.75 ))
    
    ax.set_xlabel('', fontsize=20)
    ax.set_ylabel('', fontsize=20 ,rotation=0)
    ax.xaxis.set_label_coords(1.0, 0.0, ax.get_yaxis_transform())
    ax.yaxis.set_label_coords(0.0, 1.0, ax.get_xaxis_transform())
    ax.legend(label)
    return ax 
```

```
dx = 1e-3                                                    
t1 = np.arange(0, 2, dx)                                                        #Arreglo con los valores en el tiempo que va a tomar la señal
x1 = lambda t: np.sin(2*np.pi*t)                                                #Se declara la señal como una funcón del argumento t

# Se crea la fución que permite la manipulación de la señal
def widget1(T):                                                                

    E = np.sum(x1(t1[:int(T/dx)])**2)*dx                                        #Se calcula la energia
    ax = Plot(t1, x1(t1))                                                       #Se realiza la gráfica
    ax.fill_between(t1[:int(T/dx)], x1(t1[:int(T/dx)]), alpha=0.2)
    ax.set_xlabel('t')
    ax.set_ylabel('x(t)')
    ax.set_xticks([1, 2])
    ax.set_xticklabels(['$T$', '$2T$'])
    ax.text(1.5, 0.75, r'$E_T = %.3f$' % E, fontdict={'fontsize': 15})

display(ipy.interactive(                                                        #Se crea el tablero
    widget1,
    T=ipy.FloatSlider(value=0, min=0, max=2, step=dx,
                      description='T:', continuous_update=False)))
```

## **2.2. Señales de potencia**

<p align = "justify">Una señal se clasifica como de potencia, si su potencia media $\bar{P}_x$ converge, esto es:

\begin{equation}
0< \bar{P}_x < \infty \tag{2.8}
\end{equation}

<p align = "justify">Para el caso de una señal $x(t)$  continua, su potencia media se calcula a partir de la siguiente expresión.

\begin{equation}
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \int_T |x(t)|^{2}~dt \tag{2.9}
\end{equation}

<p align = "justify">Si $x(t)$ es periódica, la ecuación anterior puede reescribirse de la siguiente manera:

\begin{equation}
\bar{P}_x = \lim_{k \to \infty} \frac{1}{k~T_0}~k \int_0^{T_0} |x(t)|^{2}~dt  \tag{2.10}
\end{equation}

<p align = "justify">Donde $T_0$ es el periodo fundamental de la señal, con lo cual se logra hacer una analogía de la ecuación inicial de potencia llevando las repeticiones al infinito, obteniendo el siguiente resultado:

\begin{equation}
\bar{P}_x = \frac{1}{T_0} \int_0^{T_0} |x(t)|^{2}~dt  \tag{2.11}
\end{equation}

<p align = "justify">Para el caso de una señal discreta $x[n]$, su potencia media está definida por:

\begin{equation}
\bar{P}_x = \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=-N}^{N} |x[n]|^{2}  \tag{2.12}
\end{equation}

<p align = "justify">De igual manera, si $x[n]$ es periódica, su potencia media estará dada por:

\begin{equation}
\bar{P}_ x = \frac{1}{N_0} \sum_{n=0}^{N_0-1} |x_[n]|^{2}  \tag{2.13}
\end{equation}
 
<p align = "justify">Donde $N_0$ es el número mínimo de muestras para el cual la señal $x[n]$ se repite. La potencia media también se puede expresar en términos de la energía de la señal de la siguiente forma:
 
* Caso continuo
 
\begin{equation}
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T}~E_T  \tag{2.14}
\end{equation}
 
* Caso discreto

\begin{equation}
\bar{P}_x = \lim_{N \to \infty} \frac{1}{2N+1}~E_N  \tag{2.15}
\end{equation}
 
De lo anterior se puede concluir que:
 
1. Si la energía de la señal es finita, entonces su potencia es cero.

2. Si la señal tiene energía infinita, su potencia puede ser finita o infinita.

3. Si la potencia de la señal es finita, su energía es infinita.

<p align = "justify">La potencia instantánea $p(t)$ para el elemento resistivo está dada por la siguiente ecuación:

$$
p(t) = v(t) \cdot i(t)  \tag{2.16}\\
 = R \cdot i(t)^{2}
$$

<p align = "justify">Evaluando en $R = 1 \Omega$.

$$
p(t) = i(t)^{2}
$$

<p align = "justify">La potencia media de la señal de corriente en un intervalo de tiempo $T$ es entonces la siguiente:

$$
\bar{P}_i = \frac{1}{T} \int_0^{T} i(t)^{2}~dt
$$

<p align = "justify">Para el caso en el que $i(t)$ coincide con una señal sinusoidal.

$$
\bar{P}_i = \frac{1}{T_0} \int_0^{T_0} \left( I_{\max}~\sin(w_0t+\theta) \right)^{2}~dt\\
\bar{P}_i = \frac{I_{\max}^{2}}{2}
$$

<p align = "justify">Lo que es equivalente al cuadrado del valor eficaz de la señal.

$$
I_{RMS} = \frac{I_{\max}}{\sqrt{2}} \tag{2.17}
$$



Código 2.2 Cálculo de la potencia media de la señal sinusoidal.

```
from sympy import *                                                             #Libreria para el uso de variables simbólicas
init_printing()                                                                 #Se adecua el espacio para imprimir de forma adecuada las expresiones simbólicas

t, I_max = symbols('t, I_max')                                                  #Se declaran las variables simbólicas
T0 = symbols('T0', nonzero = True)
omega = 2*pi / T0
x = I_max*sin(omega*t)
P = integrate(x**2, (t, 0, T0)) / T0                                            #Se calcula la potencia de la señal mediante su definición
print('La potencia media de la señal es:')                                      #Se imprime el resultado en pantalla
P
```

<p align = "justify">Código 2.3 calcula la potencia media   instantánea de una señal de corriente $i(t)=ASen(2 \pi t)$ que circula a través de una resistencia de 1 Ohmio y $A$ es un parámetro variable de $i(t)$.

```
dx = 1e-3
t2 = np.arange(0, 3, dx)

def widget2(A):

    x = A*np.sin(2*np.pi*t2)
    p = x**2
    P = np.sum(x**2)*dx * np.ones(t2.size)/3                                    #Se calcula la potencia de la señal de acuerdo a su definición
    ax = Plot(t2, x, t2, p, t2, P)
    ax.set_xlabel('t')
    ax.set_ylabel('x(t)')
    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(['$T$', '$2T$', '$3T$'])
    ax.legend(['Señal', 'Potencia instantánea', 'Potencia media'],
              loc='lower right')

display(ipy.interactive(widget2,                                                #Se crea el tablero
                        A=ipy.FloatText(value=1, step=0.1,
                                          description='A:',
                                        continuous_update=False)))
```

<p align = "justify"><b><h4>Ejemplo 2.2: <h4\></b>Para el caso de una señal sinusoidal de amplitud $A$ y periodo fundamental $T_0$ tal que:



$$
v(t) = Acos(w_0t), ~~~ w_0 = \frac{2\pi}{T_0}
$$
 
Determinar su potencia media y energía total.

**Desarrollo**

$$
\bar{P}_v = \frac{1}{T_0} \int_0^{T_0} |Acos(w_0t)|^{2}~dt\\
\bar{P}_v = \frac{A^{2}}{T_0} \left[\int_0^{T0} \frac{1}{2} ~dt + \int_0^{T0} \frac{cos(2w_0t)}{2} ~dt  \right]\\
\bar{P}_v = \frac{A^{2}}{2}
$$

<p align = "justify">Debido a que la potencia de la señal es finita (siempre y cuando $A<\infty$), su energía total es infinita.
 
$$
E_v \to \infty
$$

```
omega, A = symbols('\omega, A', nonzero = True, nonnegative = True)             #Se declaran las variables simbólicas
T0 = 2*pi/omega
x = A*cos(omega*t)
Pv = integrate(x**2, (t, 0, T0)) / T0                                            #Se calcula la potencia
print('La potencia media de la señal es:')
display(Pv)
Ev = integrate(x**2, (t, -oo, oo))                                              #Se calcula la energia
print('\nLa energía total de la señal es:')
display(Ev)
```

<p align = "justify"><b><h4>Ejemplo 2.3: <h4\></b>Para el caso de la señal $x(t) = e^{-\alpha t}u(t)$ con $\alpha > 0$, hallar su energía y potencia media.

**Desarrollo**

$$
E_x = \int_{-\infty}^{\infty} |e^{-\alpha t}u(t)|^{2}~dt\\
E_x = \int_{0}^{\infty} e^{-2\alpha t}~dt\\
E_x = -\frac{1}{2\alpha}~ e^{-2\alpha t}~\Big|_0^\infty\\
E_x = \frac{1}{2\alpha}
$$

$$
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \int_{-\frac{T}{2}}^{\frac{T}{2}} |e^{-\alpha t}u(t)|^{2}~dt\\
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \int_{0}^{\frac{T}{2}} e^{-2\alpha t}~dt\\
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \left[  -\frac{1}{2\alpha}~ e^{-2\alpha t}~\Big|_0^\frac{T}{2} \right]\\
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \left[  -\frac{1}{2\alpha}~ e^{-\alpha Tt} + \frac{1}{2\alpha} \right]\\
\bar{P}_x = 0
$$

A continuación se presenta el correspondiente código que valida los cálculos.

Código 2.5: código que valida el ejemplo 2.3

```
a = symbols('a', positive = True)                                               #Se declaran las variables simbólicas necesarias
T = symbols('T')
x = exp(-a*t)
E = integrate(x**2, (t, 0, oo))
print('La energía total de la señal es:')                                       #Se realiza la integral para calcular la potencia
display(E)
P = limit(integrate(x**2, (t, 0, T/2)) / T, T, oo)                              #Se realiza la integral para calcular la energia
print('\nLa potencia total de la señal es:')                                    
display(P)
```

<p align = "justify"><b><h4>Ejemplo 2.4: <h4\></b>Determine la potencia y energía de la señal escalón unitario discreta $\mu[n]$.

**Desarrollo**

$$
\bar{P_\mu} = \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=-N}^{N} |u[n]|^{2}\\
\bar{P_\mu} = \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=0}^{N} 1\\
\bar{P_\mu} = \lim_{N \to \infty} \frac{N+1}{2N+1}\\
\bar{P_\mu} = \frac{1}{2}
$$

Sea $x_1(t)$ y $x_2(t$) dos señales periódicas de periodo fundamental $T_0$ como se muestra a continuación.

   \right.
$$

**Desarrollo**

$$
d(x_1, x_2) = \frac{1}{T_0} \left[ \int_0^{\frac{T_0}{4}} |1-cos(w_0t)|^{2}~dt+ \int_{\frac{T_0}{4}}^{\frac{3T_0}{4}} |-1-cos(w_0t)|^{2}~dt+ \int_{\frac{3T_0}{4}}^{T_0} |1-cos(w_0t)|^{2}~dt \right]\\
