<p align = "justify">Las seÃ±ales pueden clasificarse como de energÃ­a o de potencia. Los conceptos de energÃ­a y potencia son empleados para la generalizaciÃ³n de mÃ©tricas que permitirÃ¡n tomar decisiones de convergencia en distintos tipos de anÃ¡lisis en espacios de representaciÃ³n.

## **2.1. SeÃ±ales de energÃ­a**

<p align = "justify">Se dice que una seÃ±al es de energÃ­a, si su energÃ­a total $E_x$ converge, es decir:

\begin{equation}
0< E_x < \infty \tag{2.1}
\end{equation}

<p align = "justify">Para el caso de una seÃ±al $x(t)$ contÃ­nua, su energÃ­a es calculada como:

\begin{equation}
E_x = \|x(t) \|^2 = \int_{-\infty}^{\infty} |x(t)|^{2}~dt \tag{2.2}
\end{equation}

<p align = "justify">donde la norma de la seÃ±al $x(t)$ se denota por $\|x(t) \|$. En el mismo sentido, la energÃ­a de la funciÃ³n discreta $x[n]$, es:

\begin{equation}
E_x = \|x[n] \|^2 = \sum_{n=-\infty}^{\infty} |x[n]|^{2} \tag{2.3}
\end{equation}

<p align = "justify">Si se define la energÃ­a de la seÃ±al en un intervalo de tiempo $T$ como:

\begin{equation}
E_T = \int_T |x(t)|^{2}~dt \tag{2.4}
\end{equation}

<p align = "justify">Su energÃ­a total puede expresarse en tÃ©rminos del siguiente lÃ­mite:

\begin{equation}
E_x = \lim_{T \to \infty} E_T \tag{2.5}
\end{equation}

<p align = "justify">Para las seÃ±ales discretas:

\begin{equation}
E_N = \sum_{n=-N}^{N} |x[n]|^{2} \tag{2.6}
\end{equation}

<p align = "justify">Y la energÃ­a total de $x[n]$ es:

\begin{equation}
E_x = \lim_{N \to \infty} E_N \tag{2.7}
\end{equation}


CÃ³digo 2.1: CÃ¡lculo de la energÃ­a de una seÃ±al sinusoidal.

```
# Se llaman las librerias a usar
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as ipy


# Se crea una funciÃ³n para crear grÃ¡ficos personalizados
def Plot(*arg, label=''):

    fig, ax= plt.subplots(1, figsize=(15, 8))
    ax.plot(*arg, lw=3)
    ax.spines['left'].set_position(('data',0))
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['top'].set_color('none')
    ax.set_axisbelow(False)

    for i in ax.get_xticklabels() + ax.get_yticklabels():
        i.set_fontsize(15)
        i.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.75 ))
    
    ax.set_xlabel('', fontsize=20)
    ax.set_ylabel('', fontsize=20 ,rotation=0)
    ax.xaxis.set_label_coords(1.0, 0.0, ax.get_yaxis_transform())
    ax.yaxis.set_label_coords(0.0, 1.0, ax.get_xaxis_transform())
    ax.legend(label)
    return ax 
```

```
dx = 1e-3                                                    
t1 = np.arange(0, 2, dx)                                                        #Arreglo con los valores en el tiempo que va a tomar la seÃ±al
x1 = lambda t: np.sin(2*np.pi*t)                                                #Se declara la seÃ±al como una funcÃ³n del argumento t

# Se crea la fuciÃ³n que permite la manipulaciÃ³n de la seÃ±al
def widget1(T):                                                                

    E = np.sum(x1(t1[:int(T/dx)])**2)*dx                                        #Se calcula la energia
    ax = Plot(t1, x1(t1))                                                       #Se realiza la grÃ¡fica
    ax.fill_between(t1[:int(T/dx)], x1(t1[:int(T/dx)]), alpha=0.2)
    ax.set_xlabel('t')
    ax.set_ylabel('x(t)')
    ax.set_xticks([1, 2])
    ax.set_xticklabels(['$T$', '$2T$'])
    ax.text(1.5, 0.75, r'$E_T = %.3f$' % E, fontdict={'fontsize': 15})

display(ipy.interactive(                                                        #Se crea el tablero
    widget1,
    T=ipy.FloatSlider(value=0, min=0, max=2, step=dx,
                      description='T:', continuous_update=False)))
```

## **2.2. SeÃ±ales de potencia**

<p align = "justify">Una seÃ±al se clasifica como de potencia, si su potencia media $\bar{P}_x$ converge, esto es:

\begin{equation}
0< \bar{P}_x < \infty \tag{2.8}
\end{equation}

<p align = "justify">Para el caso de una seÃ±al $x(t)$  continua, su potencia media se calcula a partir de la siguiente expresiÃ³n.

\begin{equation}
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \int_T |x(t)|^{2}~dt \tag{2.9}
\end{equation}

<p align = "justify">Si $x(t)$ es periÃ³dica, la ecuaciÃ³n anterior puede reescribirse de la siguiente manera:

\begin{equation}
\bar{P}_x = \lim_{k \to \infty} \frac{1}{k~T_0}~k \int_0^{T_0} |x(t)|^{2}~dt  \tag{2.10}
\end{equation}

<p align = "justify">Donde $T_0$ es el periodo fundamental de la seÃ±al, con lo cual se logra hacer una analogÃ­a de la ecuaciÃ³n inicial de potencia llevando las repeticiones al infinito, obteniendo el siguiente resultado:

\begin{equation}
\bar{P}_x = \frac{1}{T_0} \int_0^{T_0} |x(t)|^{2}~dt  \tag{2.11}
\end{equation}

<p align = "justify">Para el caso de una seÃ±al discreta $x[n]$, su potencia media estÃ¡ definida por:

\begin{equation}
\bar{P}_x = \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=-N}^{N} |x[n]|^{2}  \tag{2.12}
\end{equation}

<p align = "justify">De igual manera, si $x[n]$ es periÃ³dica, su potencia media estarÃ¡ dada por:

\begin{equation}
\bar{P}_ x = \frac{1}{N_0} \sum_{n=0}^{N_0-1} |x_[n]|^{2}  \tag{2.13}
\end{equation}
 
<p align = "justify">Donde $N_0$ es el nÃºmero mÃ­nimo de muestras para el cual la seÃ±al $x[n]$ se repite. La potencia media tambiÃ©n se puede expresar en tÃ©rminos de la energÃ­a de la seÃ±al de la siguiente forma:
 
* Caso continuo
 
\begin{equation}
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T}~E_T  \tag{2.14}
\end{equation}
 
* Caso discreto

\begin{equation}
\bar{P}_x = \lim_{N \to \infty} \frac{1}{2N+1}~E_N  \tag{2.15}
\end{equation}
 
De lo anterior se puede concluir que:
 
1. Si la energÃ­a de la seÃ±al es finita, entonces su potencia es cero.

2. Si la seÃ±al tiene energÃ­a infinita, su potencia puede ser finita o infinita.

3. Si la potencia de la seÃ±al es finita, su energÃ­a es infinita.

<p align = "justify">La potencia instantÃ¡nea $p(t)$ para el elemento resistivo estÃ¡ dada por la siguiente ecuaciÃ³n:

$$
p(t) = v(t) \cdot i(t)  \tag{2.16}\\
 = R \cdot i(t)^{2}
$$

<p align = "justify">Evaluando en $R = 1 \Omega$.

$$
p(t) = i(t)^{2}
$$

<p align = "justify">La potencia media de la seÃ±al de corriente en un intervalo de tiempo $T$ es entonces la siguiente:

$$
\bar{P}_i = \frac{1}{T} \int_0^{T} i(t)^{2}~dt
$$

<p align = "justify">Para el caso en el que $i(t)$ coincide con una seÃ±al sinusoidal.

$$
\bar{P}_i = \frac{1}{T_0} \int_0^{T_0} \left( I_{\max}~\sin(w_0t+\theta) \right)^{2}~dt\\
\bar{P}_i = \frac{I_{\max}^{2}}{2}
$$

<p align = "justify">Lo que es equivalente al cuadrado del valor eficaz de la seÃ±al.

$$
I_{RMS} = \frac{I_{\max}}{\sqrt{2}} \tag{2.17}
$$



CÃ³digo 2.2 CÃ¡lculo de la potencia media de la seÃ±al sinusoidal.

```
from sympy import *                                                             #Libreria para el uso de variables simbÃ³licas
init_printing()                                                                 #Se adecua el espacio para imprimir de forma adecuada las expresiones simbÃ³licas

t, I_max = symbols('t, I_max')                                                  #Se declaran las variables simbÃ³licas
T0 = symbols('T0', nonzero = True)
omega = 2*pi / T0
x = I_max*sin(omega*t)
P = integrate(x**2, (t, 0, T0)) / T0                                            #Se calcula la potencia de la seÃ±al mediante su definiciÃ³n
print('La potencia media de la seÃ±al es:')                                      #Se imprime el resultado en pantalla
P
```

<p align = "justify">CÃ³digo 2.3 calcula la potencia media   instantÃ¡nea de una seÃ±al de corriente $i(t)=ASen(2 \pi t)$ que circula a travÃ©s de una resistencia de 1 Ohmio y $A$ es un parÃ¡metro variable de $i(t)$.

```
dx = 1e-3
t2 = np.arange(0, 3, dx)

def widget2(A):

    x = A*np.sin(2*np.pi*t2)
    p = x**2
    P = np.sum(x**2)*dx * np.ones(t2.size)/3                                    #Se calcula la potencia de la seÃ±al de acuerdo a su definiciÃ³n
    ax = Plot(t2, x, t2, p, t2, P)
    ax.set_xlabel('t')
    ax.set_ylabel('x(t)')
    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(['$T$', '$2T$', '$3T$'])
    ax.legend(['SeÃ±al', 'Potencia instantÃ¡nea', 'Potencia media'],
              loc='lower right')

display(ipy.interactive(widget2,                                                #Se crea el tablero
                        A=ipy.FloatText(value=1, step=0.1,
                                          description='A:',
                                        continuous_update=False)))
```

$$

Determinar su potencia media y energÃ­a total.

<p align = "justify"><b><h4>Ejemplo 2.3: <h4\></b>Para el caso de la seÃ±al $x(t) = e^{-\alpha t}u(t)$ con $\alpha > 0$, hallar su energÃ­a y potencia media.

**Desarrollo**

$$
E_x = \int_{-\infty}^{\infty} |e^{-\alpha t}u(t)|^{2}~dt\\
E_x = \int_{0}^{\infty} e^{-2\alpha t}~dt\\
E_x = -\frac{1}{2\alpha}~ e^{-2\alpha t}~\Big|_0^\infty\\
E_x = \frac{1}{2\alpha}
$$

$$
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \int_{-\frac{T}{2}}^{\frac{T}{2}} |e^{-\alpha t}u(t)|^{2}~dt\\
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \int_{0}^{\frac{T}{2}} e^{-2\alpha t}~dt\\
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \left[  -\frac{1}{2\alpha}~ e^{-2\alpha t}~\Big|_0^\frac{T}{2} \right]\\
\bar{P}_x = \lim_{T \to \infty} \frac{1}{T} \left[  -\frac{1}{2\alpha}~ e^{-\alpha Tt} + \frac{1}{2\alpha} \right]\\
\bar{P}_x = 0
$$

A continuaciÃ³n se presenta el correspondiente cÃ³digo que valida los cÃ¡lculos.

CÃ³digo 2.5: cÃ³digo que valida el ejemplo 2.3

```
a = symbols('a', positive = True)                                               #Se declaran las variables simbÃ³licas necesarias
T = symbols('T')
x = exp(-a*t)
E = integrate(x**2, (t, 0, oo))
print('La energÃ­a total de la seÃ±al es:')                                       #Se realiza la integral para calcular la potencia
display(E)
P = limit(integrate(x**2, (t, 0, T/2)) / T, T, oo)                              #Se realiza la integral para calcular la energia
print('\nLa potencia total de la seÃ±al es:')                                    
display(P)
```
